```python
import pandas as pd

two = pd.DataFrame({'id': [1,2,3,4,5,6],'Name': ['Billy','Brian','Bran','Bryce','Betty','Dein'],'subject_id': ['sub2','sub4','sub3','sub6','sub5','sub6'],'Marks_scored': [89,80,79,97,88,99]},index = [1,2,3,4,5,6])

three = pd.DataFrame({'id': [1,2,3,4,5,6],'Name': ['Billy','Brian','Bran','Bryce','Betty','Bin'],'subject_id': ['sub2','sub4','sub3','sub6','sub5','sub7'],'Marks_scored': [89,80,79,97,88,90],'roll_no':[1,23,4,5,5,7]},index = [1,2,3,4,5,7])
```

```python
two
```

```python
three
```

```python
df = pd.merge(three,two,on = ['id'], how = 'left')

df
```

```python
df = pd.merge(three,two,on = ['id'], how = 'right')

df
```

```python
df = pd.merge(three,two,on = ['id'], how = 'outer')

df
```

```python
df = pd.merge(three,two,on = ['id'], how = 'inner')

df
```
